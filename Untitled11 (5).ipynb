```python
from bs4 import BeautifulSoup, NavigableString
import pandas as pd, re

HEADINGS = ("h1","h2","h3","h4","h5","h6")
ALLOWED  = {"p","ul","ol","table","blockquote","pre","div","section"}

def extract_outline(url, soup):
    # Prefer a main/article root if present
    root = (soup.select_one("main article") or
            soup.select_one("article") or
            soup.select_one("#main-content") or
            soup)

    def is_utility(tag):
        """True for left nav/related blocks/footers; False for main content wrappers."""
        if not tag: return False
        if tag.name in {"nav","aside","footer"}: return True
        if tag.get("role") in {"navigation","complementary"}: return True
        aria = (tag.get("aria-label") or "").lower()
        if "book outline" in aria: return True
        tid = (tag.get("id") or "")
        # left book menu block ids look like 'book-block-menu-1363' or 'block-rulebook-booknavigation'
        if tid.startswith("book-block-menu") or tid == "block-rulebook-booknavigation":
            return True
        return False

    # keep headings that are NOT inside utility containers
    hs = [h for h in root.find_all(HEADINGS) if h.find_parent(is_utility) is None]

    if not hs:
        txt = root.get_text(" ", strip=True)
        return pd.DataFrame([{
            "url": url, "order": 1, "level": 1,
            "number": None, "heading": "(unheaded content)", "text": txt
        }])

    rows = []
    for i, h in enumerate(hs, start=1):
        level = int(h.name[1])

        # collect until the next heading of same or higher level
        text_chunks = []
        for sib in h.next_siblings:
            if isinstance(sib, NavigableString):
                continue
            nm = getattr(sib, "name", None)
            if nm in HEADINGS and int(nm[1]) <= level:
                break
            if nm in ALLOWED and sib.find_parent(is_utility) is None:
                text_chunks.append(sib.get_text(" ", strip=True))
        full_text = "\n\n".join(t for t in text_chunks if t)

        title = h.get_text(" ", strip=True)
        m = re.search(r"^\s*([0-9IVXivx]+(?:\.[0-9A-Za-z]+)*)\b|\b(Pillar\s*[0-9]+)\b", title)
        number  = m.group(0) if m else None
        heading = re.sub(r"^\s*([0-9IVXivx]+(?:\.[0-9A-Za-z]+)*)\s*[-–—:]?\s*", "", title).strip()

        rows.append({
            "url": url, "order": i, "level": level,
            "number": number, "heading": heading, "text": full_text
        })

    return pd.DataFrame(rows)
```

```python
import re, json, csv, time, hashlib
from urllib.parse import urljoin, urlparse, urldefrag
from collections import deque
import requests
from bs4 import BeautifulSoup

BASE = "https://rulebook.sama.gov.sa"
LANG = "en"
START_URL = "https://rulebook.sama.gov.sa/en/prudential-and-supervisory-requirements-3/"

UA = {"User-Agent": "SAMA"}
CONTAINERS_CSS = (
    "main, article, .entry-content, .content, #content, #main, "
    ".wp-block-group, .wp-site-blocks, .wp-block-group__inner-container, .wp-block-columns"
)

def fetch(url, timeout=30):
    r = requests.get(url, headers=UA, timeout=timeout, allow_redirects=True)
    if r.status_code != 200: return None
    return BeautifulSoup(r.text, "html.parser")

def page_title(soup):
    t = soup.find("title")
    return t.get_text(" ", strip=True) if t else ""

def text_of(el):
    return re.sub(r"\s+"," ", el.get_text(" ", strip=True)).strip()

def choose_main_container(soup):
    conts = soup.select(CONTAINERS_CSS)
    return max(conts, key=lambda el: len(el.get_text().strip())) if conts else soup
```

```python
def extract_blocks(soup):
    c = choose_main_container(soup)
    blocks, seq = [], 0
    def add(kind, text, meta=None):
        nonlocal seq; seq += 1
        blocks.append({"kind": kind, "text": text, "meta": meta or {}, "block_ix": seq})

    for el in c.descendants:
        if not getattr(el, "name", None): continue
        if re.fullmatch(r"h[1-6]", el.name, re.I):
            add("heading", text_of(el), {"level": int(el.name[1])})
        elif el.name == "p":
            t = text_of(el)
            if t: add("paragraph", t)
        elif el.name in ("ul","ol"):
            items = [text_of(li) for li in el.select(":scope > li") if text_of(li)]
            if items: add("list", "\n".join(items), {"ordered": el.name=="ol", "n": len(items)})
        elif el.name == "table":
            rows = []
            for tr in el.select("tr"):
                cells = [text_of(td) for td in tr.select("th,td")]
                rows.append(cells)
            if rows:
                md = []
                if rows[0]:
                    md += ["| " + " | ".join(rows[0]) + " |",
                           "| " + " | ".join(["---"]*len(rows[0])) + " |"]
                for r in rows[1:]:
                    md.append("| " + " | ".join(r) + " |")
                add("table", "\n".join(md), {"rows": len(rows)})
    return blocks

# quick sanity check on your page
s = fetch(START_URL)
print("Title:", page_title(s))
blks = extract_blocks(s)
print("Blocks:", len(blks))
blks[:6]
```

```python
DROP_PHRASES = [
    "Entire section", "Custom print", "Text Only", "Rich Text",
    "Print / Save as PDF", "Book traversal links", "Up", "‹", "›"
]

def get_content(url):
    html = requests.get(url, timeout=30).text
    soup = BeautifulSoup(html, "html.parser")
    c = soup.select_one("main, article, .entry-content, .content, #content, .wp-block-group")
    if not c:
        return ""

    parts = []
    for node in c.select("p, h1, h2, h3, li"):
        t = node.get_text(" ", strip=True)
        if not t:
            continue
        # skip if it contains known junk
        if any(p.lower() in t.lower() for p in DROP_PHRASES):
            continue
        parts.append(t)

    text = "\n".join(parts)
    # remove pipe-heavy menu tables
    text = re.sub(r"[|]{2,}", " ", text)
    return text.strip()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
import nltk
from nltk.tokenize import word_tokenize
from typing import Dict, List, Tuple
import matplotlib.pyplot as plt
import seaborn as sns

# Download required NLTK data
nltk.download('punkt', quiet=True)

def get_content(url):
    """Your existing scraping function"""
    html = requests.get(url).text
    soup = BeautifulSoup(html, "html.parser")
    c = soup.select_one("main, article, .entry-content, .content, #content, .wp-block-group")
    if not c:
        return ""
    # --- NEW: inline placeholder for any images (equations are images on SAMA) ---
    for img in c.select("img"):
        img.replace_with(soup.new_string("[EQUATION]"))
    # ---------------------------------------------------------------------------
    parts = []
    # 1) Tables -> flattened text
    for table in c.select("table"):
        for row in table.select("tr"):
            cells = [td.get_text(" ", strip=True) for td in row.select("td, th")]
            if cells:
                parts.append(" | ".join(cells))
        parts.append("")        # blank line between tables
        table.extract()         # avoid re-reading later
    # 2) Paragraphs & headers (placeholders now appear inline)
    for el in c.select("p, h1, h2, h3, h4, h5, h6, li"):
        txt = el.get_text(" ", strip=True)
        if txt:
            parts.append(txt)
    return "\n".join(parts)


class SAMATextCleaner:
    """
    Text cleaner specifically designed for SAMA rulebook content
    Handles regulatory text peculiarities
    """
    
    def __init__(self):
        self.section_patterns = [
            r'^\d+\.\d+\.\d+',  # e.g., 3.2.1
            r'^\d+\.\d+',       # e.g., 3.2
            r'^\d+\.',          # e.g., 3.
            r'^[A-Z]\.',        # e.g., A.
            r'^\([a-z]\)',      # e.g., (a)
            r'^\([i-v]+\)',     # e.g., (i), (ii)
        ]
        
    def clean_sama_text(self, text: str) -> Dict:
        """
        Clean SAMA text and return both cleaned text and extracted components
        """
        if not text:
            return {'cleaned_text': '', 'sections': [], 'equations': 0, 'tables': 0}
        
        # Split into lines for processing
        lines = text.split('\n')
        
        cleaned_lines = []
        sections = []
        equation_count = text.count('[EQUATION]')
        table_count = text.count(' | ')  # Rough estimate based on your table format
        
        for line in lines:
            line = line.strip()
            if not line:
                continue
            
            # Check if it's a section header
            is_section = False
            for pattern in self.section_patterns:
                if re.match(pattern, line):
                    sections.append(line)
                    is_section = True
                    break
            
            # Clean the line
            # Remove excessive spaces
            line = re.sub(r'\s+', ' ', line)
            
            # Handle special characters common in regulatory text
            # Keep important punctuation for sentence detection
            line = re.sub(r'[^\w\s.,;:!?()\[\]-]', ' ', line)
            
            # Remove standalone numbers that might be page numbers or references
            if re.match(r'^\d+$', line):
                continue
            
            # Remove very short lines that are likely artifacts
            if len(line) < 10 and not is_section:
                continue
            
            cleaned_lines.append(line)
        
        cleaned_text = ' '.join(cleaned_lines)
        
        # Final cleaning passes
        # Remove multiple spaces
        cleaned_text = re.sub(r'\s+', ' ', cleaned_text)
        # Remove spaces before punctuation
        cleaned_text = re.sub(r'\s+([.,;:!?])', r'\1', cleaned_text)
        # Ensure space after punctuation
        cleaned_text = re.sub(r'([.,;:!?])([A-Za-z])', r'\1 \2', cleaned_text)
        
        return {
            'cleaned_text': cleaned_text.strip(),
            'sections': sections,
            'equations': equation_count,
            'tables': table_count
        }


class BankOfEnglandComplexityAnalyzer:
    """
    Bank of England local complexity measures
    Based on Staff Working Paper No. 834
    """
    
    def __init__(self):
        # Conditional terms from the BoE paper
        self.conditional_terms = [
            'if', 'when', 'whenever', 'where', 'wherever', 'unless',
            'notwithstanding', 'except', 'but', 'provided', 'provided that'
        ]
        
    def calculate_local_complexity(self, text: str) -> Dict:
        """
        Calculate the three local complexity measures from BoE paper
        """
        if not text:
            return {
                'tokens': 0,
                'unique_tokens': 0,
                'lexical_diversity': 0,
                'conditional_terms': 0,
                'relative_conditionality': 0
            }
        
        # Tokenization
        tokens = word_tokenize(text.lower())
        # Keep only alphanumeric tokens
        tokens = [t for t in tokens if t.isalnum()]
        
        if len(tokens) == 0:
            return {
                'tokens': 0,
                'unique_tokens': 0,
                'lexical_diversity': 0,
                'conditional_terms': 0,
                'relative_conditionality': 0
            }
        
        # 1. LENGTH
        num_tokens = len(tokens)
        
        # 2. LEXICAL DIVERSITY (Maas' index)
        unique_tokens = set(tokens)
        num_unique_tokens = len(unique_tokens)
        
        if num_tokens > 1:
            lexical_diversity = 1 - ((np.log(num_tokens) - np.log(num_unique_tokens)) / 
                                    (np.log(num_tokens) ** 2))
        else:
            lexical_diversity = 1.0
        
        # 3. CONDITIONALITY
        conditional_count = 0
        text_lower = text.lower()
        
        for term in self.conditional_terms:
            pattern = r'\b' + re.escape(term) + r'\b'
            conditional_count += len(re.findall(pattern, text_lower))
        
        relative_conditionality = conditional_count / num_tokens if num_tokens > 0 else 0
        
        return {
            'tokens': num_tokens,
            'unique_tokens': num_unique_tokens,
            'lexical_diversity': round(lexical_diversity, 4),
            'conditional_terms': conditional_count,
            'relative_conditionality': round(relative_conditionality, 4)
        }


def analyze_sama_credit_risk_section(url: str = "https://rulebook.sama.gov.sa/en/minimum-capital-requirements-credit-risk"):
    """
    Complete analysis pipeline for SAMA credit risk section
    """
    print("=" * 80)
    print("SAMA RULEBOOK - MINIMUM CAPITAL REQUIREMENTS FOR CREDIT RISK")
    print("Text Cleaning and Local Complexity Analysis")
    print("=" * 80)
    
    # Step 1: Scrape content
    print("\n1. SCRAPING CONTENT...")
    print(f"URL: {url}")
    raw_content = get_content(url)
    print(f"Raw content length: {len(raw_content)} characters")
    
    # Show sample of raw content
    print("\nFirst 500 characters of raw content:")
    print("-" * 40)
    print(raw_content[:500])
    print("...")
    
    # Step 2: Clean text
    print("\n2. CLEANING TEXT...")
    cleaner = SAMATextCleaner()
    cleaning_result = cleaner.clean_sama_text(raw_content)
    
    cleaned_text = cleaning_result['cleaned_text']
    print(f"Cleaned content length: {len(cleaned_text)} characters")
    print(f"Number of sections found: {len(cleaning_result['sections'])}")
    print(f"Number of equations: {cleaning_result['equations']}")
    print(f"Number of tables (estimated): {cleaning_result['tables']}")
    
    # Show sample of cleaned content
    print("\nFirst 500 characters of cleaned content:")
    print("-" * 40)
    print(cleaned_text[:500])
    print("...")
    
    # Show some extracted sections
    if cleaning_result['sections']:
        print("\nSample section headers found:")
        print("-" * 40)
        for section in cleaning_result['sections'][:10]:
            print(f"  • {section}")
        if len(cleaning_result['sections']) > 10:
            print(f"  ... and {len(cleaning_result['sections']) - 10} more")
    
    # Step 3: Calculate complexity measures
    print("\n3. CALCULATING LOCAL COMPLEXITY MEASURES...")
    analyzer = BankOfEnglandComplexityAnalyzer()
    complexity_measures = analyzer.calculate_local_complexity(cleaned_text)
    
    print("\nLocal Complexity Measures (Bank of England methodology):")
    print("-" * 40)
    print(f"Number of tokens (words):        {complexity_measures['tokens']:,}")
    print(f"Number of unique tokens:          {complexity_measures['unique_tokens']:,}")
    print(f"Lexical diversity (Maas' index):  {complexity_measures['lexical_diversity']:.4f}")
    print(f"Conditional terms count:          {complexity_measures['conditional_terms']}")
    print(f"Relative conditionality:          {complexity_measures['relative_conditionality']:.4f}")
    
    # Step 4: Break down by sections for more detailed analysis
    print("\n4. ANALYZING TEXT BY SECTIONS...")
    
    # Split text into paragraphs/sections for granular analysis
    paragraphs = [p.strip() for p in cleaned_text.split('.') if len(p.strip()) > 50]
    
    if paragraphs:
        section_complexities = []
        for i, paragraph in enumerate(paragraphs[:20], 1):  # Analyze first 20 paragraphs
            para_complexity = analyzer.calculate_local_complexity(paragraph)
            section_complexities.append({
                'section': i,
                'tokens': para_complexity['tokens'],
                'lexical_diversity': para_complexity['lexical_diversity'],
                'conditional_terms': para_complexity['conditional_terms']
            })
        
        df_sections = pd.DataFrame(section_complexities)
        
        print(f"\nComplexity of first {len(df_sections)} text segments:")
        print("-" * 40)
        print(df_sections.describe().round(3))
        
        # Find most complex segments
        most_complex_by_tokens = df_sections.nlargest(3, 'tokens')
        print("\nMost complex segments by length (tokens):")
        print(most_complex_by_tokens[['section', 'tokens', 'lexical_diversity']])
        
        if df_sections['conditional_terms'].sum() > 0:
            most_conditional = df_sections[df_sections['conditional_terms'] > 0]
            print(f"\nSegments with conditional terms: {len(most_conditional)}")
            print(most_conditional[['section', 'conditional_terms']].head())
    
    # Step 5: Create visualization
    print("\n5. CREATING VISUALIZATION...")
    
    # Create a summary visualization
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    
    # Plot 1: Word frequency distribution
    ax1 = axes[0, 0]
    words = word_tokenize(cleaned_text.lower())
    words = [w for w in words if w.isalnum() and len(w) > 3]
    word_freq = pd.Series(words).value_counts().head(20)
    ax1.barh(range(len(word_freq)), word_freq.values)
    ax1.set_yticks(range(len(word_freq)))
    ax1.set_yticklabels(word_freq.index)
    ax1.set_xlabel('Frequency')
    ax1.set_title('Top 20 Most Frequent Words')
    ax1.invert_yaxis()
    
    # Plot 2: Complexity measures summary
    ax2 = axes[0, 1]
    measures = ['Tokens', 'Unique\nTokens', 'Conditional\nTerms']
    values = [
        complexity_measures['tokens'],
        complexity_measures['unique_tokens'],
        complexity_measures['conditional_terms'] * 100  # Scale for visibility
    ]
    bars = ax2.bar(measures, values, color=['#1f77b4', '#ff7f0e', '#2ca02c'])
    ax2.set_ylabel('Count')
    ax2.set_title('Local Complexity Measures')
    # Add value labels on bars
    for bar, val in zip(bars, [complexity_measures['tokens'], 
                               complexity_measures['unique_tokens'],
                               complexity_measures['conditional_terms']]):
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height,
                f'{int(val):,}', ha='center', va='bottom')
    
    # Plot 3: Lexical diversity and conditionality
    ax3 = axes[1, 0]
    metrics = ['Lexical\nDiversity', 'Relative\nConditionality']
    metric_values = [
        complexity_measures['lexical_diversity'],
        complexity_measures['relative_conditionality'] * 10  # Scale for visibility
    ]
    bars = ax3.bar(metrics, metric_values, color=['#d62728', '#9467bd'])
    ax3.set_ylabel('Score')
    ax3.set_title('Normalized Complexity Metrics')
    # Add value labels
    for bar, val in zip(bars, [complexity_measures['lexical_diversity'],
                               complexity_measures['relative_conditionality']]):
        height = bar.get_height()
        ax3.text(bar.get_x() + bar.get_width()/2., height,
                f'{val:.4f}', ha='center', va='bottom')
    
    # Plot 4: Section complexity distribution (if we have sections)
    ax4 = axes[1, 1]
    if 'df_sections' in locals() and not df_sections.empty:
        ax4.plot(df_sections['section'], df_sections['tokens'], 'o-', label='Tokens per segment')
        ax4.set_xlabel('Text Segment')
        ax4.set_ylabel('Number of Tokens')
        ax4.set_title('Complexity Distribution Across Text Segments')
        ax4.grid(True, alpha=0.3)
        ax4.legend()
    else:
        ax4.text(0.5, 0.5, 'No section data available', 
                ha='center', va='center', transform=ax4.transAxes)
        ax4.set_title('Section Analysis')
    
    plt.suptitle('SAMA Credit Risk Requirements - Text Complexity Analysis', 
                fontsize=14, fontweight='bold')
    plt.tight_layout()
    
    # Return results
    return {
        'url': url,
        'raw_content': raw_content,
        'cleaned_text': cleaned_text,
        'cleaning_metadata': cleaning_result,
        'complexity_measures': complexity_measures,
        'figure': fig
    }


# Run the analysis
if __name__ == "__main__":
    # Analyze the credit risk section
    results = analyze_sama_credit_risk_section()
    
    # Save cleaned text to file
    with open('sama_credit_risk_cleaned.txt', 'w', encoding='utf-8') as f:
        f.write(results['cleaned_text'])
    print("\n✓ Cleaned text saved to 'sama_credit_risk_cleaned.txt'")
    
    # Save complexity measures to CSV
    measures_df = pd.DataFrame([results['complexity_measures']])
    measures_df['url'] = results['url']
    measures_df['raw_length'] = len(results['raw_content'])
    measures_df['cleaned_length'] = len(results['cleaned_text'])
    measures_df.to_csv('sama_credit_risk_complexity.csv', index=False)
    print("✓ Complexity measures saved to 'sama_credit_risk_complexity.csv'")
    
    # Show the visualization
    plt.show()
    
    print("\n" + "=" * 80)
    print("ANALYSIS COMPLETE!")
    print("=" * 80)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from typing import Dict, List, Tuple
import matplotlib.pyplot as plt
# seaborn imported earlier but not used; fine to keep/remove.

# Download required NLTK data
nltk.download('punkt', quiet=True)

def get_content(url):
    """Scrape page main content, flatten tables, mark equations and tables."""
    resp = requests.get(
        url,
        headers={"User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 "
                               "(KHTML, like Gecko) Chrome/120.0 Safari/537.36"},
        timeout=20
    )
    resp.raise_for_status()
    html = resp.text
    soup = BeautifulSoup(html, "html.parser")
    c = soup.select_one("main, article, .entry-content, .content, #content, .wp-block-group")
    if not c:
        return ""

    # Inline placeholder for images (SAMA often renders equations as images)
    for img in c.select("img"):
        img.replace_with(soup.new_string("[EQUATION]"))

    parts = []

    # 1) Tables -> flattened text, with a marker per table
    for table in c.select("table"):
        # Mark this table once so we can count tables precisely downstream
        parts.append("[TABLE]")

        for row in table.select("tr"):
            cells = [td.get_text(" ", strip=True) for td in row.select("td, th")]
            if cells:
                parts.append(" | ".join(cells))
        parts.append("")        # blank line between tables
        table.extract()         # avoid re-reading later

    # 2) Paragraphs & headers (placeholders now appear inline)
    for el in c.select("p, h1, h2, h3, h4, h5, h6, li"):
        txt = el.get_text(" ", strip=True)
        if txt:
            parts.append(txt)

    return "\n".join(parts)


class SAMATextCleaner:
    """
    Text cleaner specifically designed for SAMA rulebook content
    Handles regulatory text peculiarities
    """

    def __init__(self):
        self.section_patterns = [
            r'^\d+\.\d+\.\d+',   # e.g., 3.2.1
            r'^\d+\.\d+',        # e.g., 3.2
            r'^\d+\.',           # e.g., 3.
            r'^[A-Z]\.',         # e.g., A.
            r'^\([a-z]\)',       # e.g., (a)
            r'^\([ivxlcdm]+\)',  # e.g., (i), (ii), (xiv) ... (broader roman numerals)
        ]

    def clean_sama_text(self, text: str) -> Dict:
        """
        Clean SAMA text and return both cleaned text and extracted components
        """
        if not text:
            return {'cleaned_text': '', 'sections': [], 'equations': 0, 'tables': 0}

        # Split into lines for processing
        lines = text.split('\n')

        cleaned_lines: List[str] = []
        sections: List[str] = []

        # Exact placeholder counts
        equation_count = text.count('[EQUATION]')
        table_count = text.count('[TABLE]')  # precise table count

        for line in lines:
            line = line.strip()
            if not line:
                continue

            # Check if it's a section header
            is_section = False
            for pattern in self.section_patterns:
                if re.match(pattern, line):
                    sections.append(line)
                    is_section = True
                    break

            # Clean the line
            # Remove excessive spaces
            line = re.sub(r'\s+', ' ', line)

            # Handle special characters common in regulatory text
            # Keep important punctuation for sentence detection
            line = re.sub(r'[^\w\s.,;:!?()\[\]-]', ' ', line)

            # Remove standalone numbers that might be page numbers or references
            if re.match(r'^\d+$', line):
                continue

            # Remove very short lines that are likely artifacts
            if len(line) < 10 and not is_section:
                continue

            cleaned_lines.append(line)

        cleaned_text = ' '.join(cleaned_lines)

        # Final cleaning passes
        cleaned_text = re.sub(r'\s+', ' ', cleaned_text)                     # collapse spaces
        cleaned_text = re.sub(r'\s+([.,;:!?])', r'\1', cleaned_text)         # no space before punct
        cleaned_text = re.sub(r'([.,;:!?])([A-Za-z])', r'\1 \2', cleaned_text)  # space after punct

        return {
            'cleaned_text': cleaned_text.strip(),
            'sections': sections,
            'equations': equation_count,
            'tables': table_count
        }


class BankOfEnglandComplexityAnalyzer:
    """
    Bank of England local complexity measures
    Based on Staff Working Paper No. 834
    """

    def __init__(self):
        # Conditional terms from the BoE paper
        self.conditional_terms = [
            'provided that',  # longer phrase first to avoid overlap
            'if', 'when', 'whenever', 'where', 'wherever', 'unless',
            'notwithstanding', 'except', 'but', 'provided'
        ]

    def calculate_local_complexity(self, text: str) -> Dict:
        """
        Calculate the three local complexity measures from BoE paper
        """
        if not text:
            return {
                'tokens': 0,
                'unique_tokens': 0,
                'lexical_diversity': 0,
                'conditional_terms': 0,
                'relative_conditionality': 0
            }

        # Tokenization
        tokens = word_tokenize(text.lower())
        # Keep only alphanumeric tokens
        tokens = [t for t in tokens if t.isalnum()]

        if len(tokens) == 0:
            return {
                'tokens': 0,
                'unique_tokens': 0,
                'lexical_diversity': 0,
                'conditional_terms': 0,
                'relative_conditionality': 0
            }

        # 1. LENGTH
        num_tokens = len(tokens)

        # 2. LEXICAL DIVERSITY (Maas' index, normalized to ~0–1)
        unique_tokens = set(tokens)
        num_unique_tokens = len(unique_tokens)

        if num_tokens > 1:
            lexical_diversity = 1 - ((np.log(num_tokens) - np.log(num_unique_tokens)) /
                                     (np.log(num_tokens) ** 2))
        else:
            lexical_diversity = 1.0

        # 3. CONDITIONALITY (avoid double-counting overlaps)
        text_lower = text.lower()
        conditional_count = 0
        # consume matches as we go so shorter terms don't re-match the same span
        for term in sorted(self.conditional_terms, key=len, reverse=True):
            pattern = r'\b' + re.escape(term) + r'\b'
            matches = list(re.finditer(pattern, text_lower))
            conditional_count += len(matches)
            # replace matched spans with spaces to avoid overlapping counts
            text_lower = re.sub(pattern, ' ', text_lower)

        relative_conditionality = conditional_count / num_tokens if num_tokens > 0 else 0

        return {
            'tokens': num_tokens,
            'unique_tokens': num_unique_tokens,
            'lexical_diversity': round(lexical_diversity, 4),
            'conditional_terms': conditional_count,
            'relative_conditionality': round(relative_conditionality, 4)
        }


def analyze_sama_credit_risk_section(url: str = "https://rulebook.sama.gov.sa/en/minimum-capital-requirements-credit-risk"):
    """
    Complete analysis pipeline for SAMA credit risk section
    """
    print("=" * 80)
    print("SAMA RULEBOOK - MINIMUM CAPITAL REQUIREMENTS FOR CREDIT RISK")
    print("Text Cleaning and Local Complexity Analysis")
    print("=" * 80)

    # Step 1: Scrape content
    print("\n1. SCRAPING CONTENT...")
    print(f"URL: {url}")
    raw_content = get_content(url)
    print(f"Raw content length: {len(raw_content)} characters")

    # Show sample of raw content
    print("\nFirst 500 characters of raw content:")
    print("-" * 40)
    print(raw_content[:500])
    print("...")

    # Step 2: Clean text
    print("\n2. CLEANING TEXT...")
    cleaner = SAMATextCleaner()
    cleaning_result = cleaner.clean_sama_text(raw_content)

    cleaned_text = cleaning_result['cleaned_text']
    print(f"Cleaned content length: {len(cleaned_text)} characters")
    print(f"Number of sections found: {len(cleaning_result['sections'])}")
    print(f"Number of equations: {cleaning_result['equations']}")
    print(f"Number of tables: {cleaning_result['tables']}")  # now exact

    # Show sample of cleaned content
    print("\nFirst 500 characters of cleaned content:")
    print("-" * 40)
    print(cleaned_text[:500])
    print("...")

    # Show some extracted sections
    if cleaning_result['sections']:
        print("\nSample section headers found:")
        print("-" * 40)
        for section in cleaning_result['sections'][:10]:
            print(f"  • {section}")
        if len(cleaning_result['sections']) > 10:
            print(f"  ... and {len(cleaning_result['sections']) - 10} more")

    # Step 3: Calculate complexity measures
    print("\n3. CALCULATING LOCAL COMPLEXITY MEASURES...")
    analyzer = BankOfEnglandComplexityAnalyzer()
    complexity_measures = analyzer.calculate_local_complexity(cleaned_text)

    print("\nLocal Complexity Measures (Bank of England methodology):")
    print("-" * 40)
    print(f"Number of tokens (words):        {complexity_measures['tokens']:,}")
    print(f"Number of unique tokens:          {complexity_measures['unique_tokens']:,}")
    print(f"Lexical diversity (Maas' index):  {complexity_measures['lexical_diversity']:.4f}")
    print(f"Conditional terms count:          {complexity_measures['conditional_terms']}")
    print(f"Relative conditionality:          {complexity_measures['relative_conditionality']:.4f}")

    # Step 4: Break down by sections for more detailed analysis
    print("\n4. ANALYZING TEXT BY SECTIONS...")

    # Use sentence tokenizer (better than splitting on '.')
    sentences = [s.strip() for s in sent_tokenize(cleaned_text) if len(s.strip()) > 50]

    if sentences:
        section_complexities = []
        for i, paragraph in enumerate(sentences[:20], 1):  # Analyze first 20 segments
            para_complexity = analyzer.calculate_local_complexity(paragraph)
            section_complexities.append({
                'section': i,
                'tokens': para_complexity['tokens'],
                'lexical_diversity': para_complexity['lexical_diversity'],
                'conditional_terms': para_complexity['conditional_terms']
            })

        df_sections = pd.DataFrame(section_complexities)

        print(f"\nComplexity of first {len(df_sections)} text segments:")
        print("-" * 40)
        print(df_sections.describe().round(3))

        # Find most complex segments
        most_complex_by_tokens = df_sections.nlargest(3, 'tokens')
        print("\nMost complex segments by length (tokens):")
        print(most_complex_by_tokens[['section', 'tokens', 'lexical_diversity']])

        if df_sections['conditional_terms'].sum() > 0:
            most_conditional = df_sections[df_sections['conditional_terms'] > 0]
            print(f"\nSegments with conditional terms: {len(most_conditional)}")
            print(most_conditional[['section', 'conditional_terms']].head())

    # Step 5: Create visualization
    print("\n5. CREATING VISUALIZATION...")

    fig, axes = plt.subplots(2, 2, figsize=(12, 10))

    # Plot 1: Word frequency distribution
    ax1 = axes[0, 0]
    words = word_tokenize(cleaned_text.lower())
    words = [w for w in words if w.isalnum() and len(w) > 3]
    word_freq = pd.Series(words).value_counts().head(20)
    ax1.barh(range(len(word_freq)), word_freq.values)
    ax1.set_yticks(range(len(word_freq)))
    ax1.set_yticklabels(word_freq.index)
    ax1.set_xlabel('Frequency')
    ax1.set_title('Top 20 Most Frequent Words')
    ax1.invert_yaxis()

    # Plot 2: Complexity measures summary
    ax2 = axes[0, 1]
    measures = ['Tokens', 'Unique\nTokens', 'Conditional\nTerms']
    values = [
        complexity_measures['tokens'],
        complexity_measures['unique_tokens'],
        complexity_measures['conditional_terms']
    ]
    bars = ax2.bar(measures, values)
    ax2.set_ylabel('Count')
    ax2.set_title('Local Complexity Measures')
    for bar, val in zip(bars, values):
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height, f'{int(val):,}',
                 ha='center', va='bottom')

    # Plot 3: Lexical diversity and conditionality
    ax3 = axes[1, 0]
    metrics = ['Lexical\nDiversity', 'Relative\nConditionality']
    metric_values = [
        complexity_measures['lexical_diversity'],
        complexity_measures['relative_conditionality']
    ]
    bars = ax3.bar(metrics, metric_values)
    ax3.set_ylabel('Score')
    ax3.set_title('Normalized Complexity Metrics')
    for bar, val in zip(bars, metric_values):
        height = bar.get_height()
        ax3.text(bar.get_x() + bar.get_width()/2., height, f'{val:.4f}',
                 ha='center', va='bottom')

    # Plot 4: Section complexity distribution (if we have sections)
    ax4 = axes[1, 1]
    if 'df_sections' in locals() and not df_sections.empty:
        ax4.plot(df_sections['section'], df_sections['tokens'], 'o-', label='Tokens per segment')
        ax4.set_xlabel('Text Segment')
        ax4.set_ylabel('Number of Tokens')
        ax4.set_title('Complexity Distribution Across Text Segments')
        ax4.grid(True, alpha=0.3)
        ax4.legend()
    else:
        ax4.text(0.5, 0.5, 'No section data available',
                 ha='center', va='center', transform=ax4.transAxes)
        ax4.set_title('Section Analysis')

    plt.suptitle('SAMA Credit Risk Requirements - Text Complexity Analysis',
                 fontsize=14, fontweight='bold')
    plt.tight_layout(rect=[0, 0, 1, 0.96])

    # Return results
    return {
        'url': url,
        'raw_content': raw_content,
        'cleaned_text': cleaned_text,
        'cleaning_metadata': cleaning_result,
        'complexity_measures': complexity_measures,
        'figure': fig
    }


# Run the analysis
if __name__ == "__main__":
    results = analyze_sama_credit_risk_section()

    # Save cleaned text to file
    with open('sama_credit_risk_cleaned.txt', 'w', encoding='utf-8') as f:
        f.write(results['cleaned_text'])
    print("\n✓ Cleaned text saved to 'sama_credit_risk_cleaned.txt'")

    # Save complexity measures to CSV
    measures_df = pd.DataFrame([results['complexity_measures']])
    measures_df['url'] = results['url']
    measures_df['raw_length'] = len(results['raw_content'])
    measures_df['cleaned_length'] = len(results['cleaned_text'])
    measures_df.to_csv('sama_credit_risk_complexity.csv', index=False)
    print("✓ Complexity measures saved to 'sama_credit_risk_complexity.csv'")

    # Show the visualization
    plt.show()

    print("\n" + "=" * 80)
    print("ANALYSIS COMPLETE!")
    print("=" * 80)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from typing import Dict, List
import matplotlib.pyplot as plt

# NLTK data
nltk.download("punkt", quiet=True)

# ----------------------- SCRAPER -----------------------
def get_content(url: str) -> str:
    """Scrape page main content, flatten tables, and mark equations/tables."""
    resp = requests.get(
        url,
        headers={"User-Agent": "Mozilla/5.0"},
        timeout=20
    )
    resp.raise_for_status()
    soup = BeautifulSoup(resp.text, "html.parser")
    c = soup.select_one("main, article, .entry-content, .content, #content, .wp-block-group")
    if not c:
        return ""

    # Mark equations (often rendered as images on SAMA)
    for img in c.select("img"):
        img.replace_with(soup.new_string("[EQUATION]"))

    parts: List[str] = []

    # Tables -> flattened text + one [TABLE] marker per table
    for table in c.select("table"):
        parts.append("[TABLE]")  # exact table counter
        for row in table.select("tr"):
            cells = [td.get_text(" ", strip=True) for td in row.select("td, th")]
            if cells:
                parts.append(" | ".join(cells))
        parts.append("")         # blank line between tables
        table.extract()

    # Paragraphs & headers
    for el in c.select("p, h1, h2, h3, h4, h5, h6, li"):
        txt = el.get_text(" ", strip=True)
        if txt:
            parts.append(txt)

    return "\n".join(parts)


# ----------------------- CLEANER (SECTION-AWARE) -----------------------
class SAMATextCleaner:
    """
    Cleans SAMA rulebook text and returns:
      - cleaned_text (flat string)
      - sections (list of section titles/headers)
      - section_blocks (list of {'title', 'text'})  <-- segments that match the rulebook
      - counts for equations and tables
    """

    def __init__(self):
        self.section_patterns = [
            r"^\d+\.\d+\.\d+",   # 3.2.1
            r"^\d+\.\d+",        # 3.2
            r"^\d+\.",           # 3.
            r"^[A-Z]\.",         # A.
            r"^\([a-z]\)",       # (a)
            r"^\([ivxlcdm]+\)",  # (i), (ii), (xiv) ...
        ]

    def _is_section_header(self, line: str) -> bool:
        for p in self.section_patterns:
            if re.match(p, line):
                return True
        return False

    def clean_sama_text(self, text: str) -> Dict:
        if not text:
            return {"cleaned_text": "", "sections": [], "section_blocks": [],
                    "equations": 0, "tables": 0}

        lines = [ln.strip() for ln in text.split("\n")]

        cleaned_lines: List[str] = []
        sections: List[str] = []
        section_blocks: List[Dict[str, str]] = []
        current_title = "Preamble"
        current_buf: List[str] = []

        equation_count = text.count("[EQUATION]")
        table_count = text.count("[TABLE]")

        for raw in lines:
            if not raw:
                continue

            is_section = self._is_section_header(raw)

            # Normalize/clean the line
            line = re.sub(r"\s+", " ", raw)
            line = re.sub(r"[^\w\s.,;:!?()\[\]-]", " ", line)

            # drop standalone numbers
            if re.match(r"^\d+$", line):
                continue
            # drop tiny artifacts unless it's a header
            if len(line) < 10 and not is_section:
                continue

            # --- build section-aligned blocks ---
            if is_section:
                # flush previous block
                if current_buf:
                    section_blocks.append({"title": current_title, "text": " ".join(current_buf)})
                    current_buf = []
                current_title = raw
                sections.append(raw)
            else:
                current_buf.append(line)

            # keep a global cleaned stream as well
            cleaned_lines.append(line if not is_section else raw + " " + line)

        # flush last
        if current_buf:
            section_blocks.append({"title": current_title, "text": " ".join(current_buf)})

        cleaned_text = " ".join(cleaned_lines)
        cleaned_text = re.sub(r"\s+", " ", cleaned_text)
        cleaned_text = re.sub(r"\s+([.,;:!?])", r"\1", cleaned_text)
        cleaned_text = re.sub(r"([.,;:!?])([A-Za-z])", r"\1 \2", cleaned_text)

        return {
            "cleaned_text": cleaned_text.strip(),
            "sections": sections,
            "section_blocks": section_blocks,
            "equations": equation_count,
            "tables": table_count
        }


# ----------------------- COMPLEXITY ANALYZER -----------------------
class BankOfEnglandComplexityAnalyzer:
    """Local complexity measures inspired by BoE SWP No. 834."""

    def __init__(self):
        # Longer phrase first to avoid overlap with "provided"
        self.conditional_terms = [
            "provided that",
            "if", "when", "whenever", "where", "wherever", "unless",
            "notwithstanding", "except", "but", "provided",
        ]

    def calculate_local_complexity(self, text: str) -> Dict:
        if not text:
            return {"tokens": 0, "unique_tokens": 0, "lexical_diversity": 0,
                    "conditional_terms": 0, "relative_conditionality": 0}

        tokens = [t for t in word_tokenize(text.lower()) if t.isalnum()]
        if not tokens:
            return {"tokens": 0, "unique_tokens": 0, "lexical_diversity": 0,
                    "conditional_terms": 0, "relative_conditionality": 0}

        num_tokens = len(tokens)
        num_unique = len(set(tokens))

        if num_tokens > 1:
            lex_div = 1 - ((np.log(num_tokens) - np.log(num_unique)) / (np.log(num_tokens) ** 2))
        else:
            lex_div = 1.0

        # Count conditionals without double-counting overlaps
        t = text.lower()
        cond_count = 0
        for term in sorted(self.conditional_terms, key=len, reverse=True):
            pat = r"\b" + re.escape(term) + r"\b"
            matches = list(re.finditer(pat, t))
            cond_count += len(matches)
            t = re.sub(pat, " ", t)

        rel_cond = cond_count / num_tokens

        return {
            "tokens": num_tokens,
            "unique_tokens": num_unique,
            "lexical_diversity": round(lex_div, 4),
            "conditional_terms": cond_count,
            "relative_conditionality": round(rel_cond, 4),
        }


# ----------------------- PIPELINE (SECTION-ALIGNED) -----------------------
def analyze_sama_credit_risk_section(
    url: str = "https://rulebook.sama.gov.sa/en/minimum-capital-requirements-credit-risk"
):
    print("=" * 80)
    print("SAMA RULEBOOK - MINIMUM CAPITAL REQUIREMENTS FOR CREDIT RISK")
    print("Text Cleaning and Local Complexity Analysis (SECTION-ALIGNED)")
    print("=" * 80)

    # 1) Scrape
    print("\n1. SCRAPING CONTENT...")
    print(f"URL: {url}")
    raw = get_content(url)
    print(f"Raw content length: {len(raw)} characters")

    print("\nFirst 500 characters of raw content:")
    print("-" * 40)
    print(raw[:500])
    print("...")

    # 2) Clean (and segment by sections)
    print("\n2. CLEANING + SECTION SPLIT...")
    cleaner = SAMATextCleaner()
    meta = cleaner.clean_sama_text(raw)
    cleaned = meta["cleaned_text"]
    blocks = meta["section_blocks"]

    print(f"Cleaned content length: {len(cleaned)} characters")
    print(f"Sections detected: {len(meta['sections'])}")
    print(f"Equations (placeholders): {meta['equations']}")
    print(f"Tables (exact): {meta['tables']}")

    if meta["sections"]:
        print("\nSample section headers:")
        for s in meta["sections"][:10]:
            print("  •", s)
        if len(meta["sections"]) > 10:
            print(f"  ... and {len(meta['sections']) - 10} more")

    # 3) Global complexity
    print("\n3. GLOBAL COMPLEXITY (BoE-style)...")
    analyzer = BankOfEnglandComplexityAnalyzer()
    global_meas = analyzer.calculate_local_complexity(cleaned)
    for k, v in [
        ("Number of tokens (words)", f"{global_meas['tokens']:,}"),
        ("Number of unique tokens", f"{global_meas['unique_tokens']:,}"),
        ("Lexical diversity (Maas')", f"{global_meas['lexical_diversity']:.4f}"),
        ("Conditional terms", f"{global_meas['conditional_terms']}"),
        ("Relative conditionality", f"{global_meas['relative_conditionality']:.4f}"),
    ]:
        print(f"{k:30s}: {v}")

    # 4) Per-section complexity (segments now match rulebook sections)
    print("\n4. PER-SECTION COMPLEXITY...")
    section_rows = []
    for i, b in enumerate(blocks, 1):
        cm = analyzer.calculate_local_complexity(b["text"])
        section_rows.append({
            "idx": i,
            "title": b["title"],
            "tokens": cm["tokens"],
            "lexical_diversity": cm["lexical_diversity"],
            "conditional_terms": cm["conditional_terms"],
        })
    df_sections = pd.DataFrame(section_rows)
    if not df_sections.empty:
        print(df_sections[["idx", "tokens", "lexical_diversity", "conditional_terms"]].head())

    # 5) Visualization
    print("\n5. VISUALIZATION...")
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))

    # Top words
    ax1 = axes[0, 0]
    words = [w for w in word_tokenize(cleaned.lower()) if w.isalnum() and len(w) > 3]
    wf = pd.Series(words).value_counts().head(20)
    ax1.barh(range(len(wf)), wf.values)
    ax1.set_yticks(range(len(wf)))
    ax1.set_yticklabels(wf.index)
    ax1.set_title("Top 20 Most Frequent Words")
    ax1.set_xlabel("Frequency")
    ax1.invert_yaxis()

    # Global measures
    ax2 = axes[0, 1]
    labels = ["Tokens", "Unique\nTokens", "Conditional\nTerms"]
    vals = [global_meas["tokens"], global_meas["unique_tokens"], global_meas["conditional_terms"]]
    bars = ax2.bar(labels, vals)
    ax2.set_title("Local Complexity Measures")
    ax2.set_ylabel("Count")
    for b, v in zip(bars, vals):
        ax2.text(b.get_x() + b.get_width() / 2, b.get_height(), f"{int(v):,}", ha="center", va="bottom")

    # Normalized metrics
    ax3 = axes[1, 0]
    nlabels = ["Lexical\nDiversity", "Relative\nConditionality"]
    nvals = [global_meas["lexical_diversity"], global_meas["relative_conditionality"]]
    nbars = ax3.bar(nlabels, nvals)
    ax3.set_title("Normalized Complexity Metrics")
    ax3.set_ylabel("Score")
    for b, v in zip(nbars, nvals):
        ax3.text(b.get_x() + b.get_width() / 2, b.get_height(), f"{v:.4f}", ha="center", va="bottom")

    # Per-section tokens (first 40 for legibility)
    ax4 = axes[1, 1]
    if not df_sections.empty:
        plot_df = df_sections.iloc[:40]
        ax4.plot(plot_df["idx"], plot_df["tokens"], "o-", label="Tokens per section")
        ax4.set_xlabel("Section index")
        ax4.set_ylabel("Tokens")
        ax4.set_title("Complexity Across Sections (aligned to headers)")
        ax4.grid(True, alpha=0.3)
        ax4.legend()
    else:
        ax4.text(0.5, 0.5, "No section data", ha="center", va="center", transform=ax4.transAxes)

    plt.suptitle("SAMA Credit Risk Requirements - Text Complexity Analysis (Section-Aligned)",
                 fontsize=14, fontweight="bold")
    plt.tight_layout(rect=[0, 0, 1, 0.96])

    return {
        "url": url,
        "raw_content": raw,
        "cleaned_text": cleaned,
        "cleaning_metadata": meta,
        "sections_df": df_sections,
        "complexity_measures": global_meas,
        "figure": fig,
    }


# ----------------------- CLI -----------------------
if __name__ == "__main__":
    results = analyze_sama_credit_risk_section()
    with open("sama_credit_risk_cleaned.txt", "w", encoding="utf-8") as f:
        f.write(results["cleaned_text"])
    print("\n✓ Cleaned text saved to 'sama_credit_risk_cleaned.txt'")

    measures_df = pd.DataFrame([results["complexity_measures"]])
    measures_df["url"] = results["url"]
    measures_df["raw_length"] = len(results["raw_content"])
    measures_df["cleaned_length"] = len(results["cleaned_text"])
    measures_df.to_csv("sama_credit_risk_complexity.csv", index=False)
    print("✓ Complexity measures saved to 'sama_credit_risk_complexity.csv'")

    plt.show()
    print("\n" + "=" * 80)
    print("ANALYSIS COMPLETE!")
    print("=" * 80)
```
